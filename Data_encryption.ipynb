```python
def ceaser_cipher(text, key):
    result = ''
    for i in text:
        if i.isupper():
            result += chr((ord(i) - 65 + key) % 26 + 65)
        elif i.islower():
            result += chr((ord(i) - 97 + key) % 26 + 97)
        else:
            result += i
    return result

# Correct usage with DataFrame
df['Username_encrypted'] = df['Username'].apply(lambda x: ceaser_cipher(x, 3))
```
